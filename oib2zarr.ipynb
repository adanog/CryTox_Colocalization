## Load libraries

```python
import os
from oiffile import imread
import matplotlib.pyplot as plt
import zarr
```

## Parameters

```python
directory_path = '/home/jovyan/LNMA/bravoa/data/New2 Fig para colocalizacion Manders-Mito'
filename = '1-Cry11 2mg 2h MitoT 750nM 60xmed-post 3.5 Z copy.oib'
file_path = os.path.join(directory_path, filename)
print(file_path)
```

## Load *.oib file

```python
image = imread(file_path)
image.shape
plt.imshow(image[3,:,:])
```

## Create *.zarr file

```python
zarr_file_path = f'{file_path[:-4]}.zarr'
zarr.save(zarr_file_path, image)
loaded_image = zarr.load(zarr_file_path)
plt.imshow(loaded_image[3,:,:])
```
