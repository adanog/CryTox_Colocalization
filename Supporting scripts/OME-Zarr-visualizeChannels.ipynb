## Import libraries

```python
import zarr
import os
import matplotlib.pyplot as plt
import numpy as np
```

## Define File Paths

```python
directory_path = '/home/jovyan/LNMA/bravoa/data/New2 Fig para colocalizacion Manders-Mito'
ome_zarr_name = '1-Cry11 2mg 2h MitoT 750nM 60xmed-post 3.5 Z copy.zarr'
ome_zarr_path = os.path.join(directory_path, ome_zarr_name)
ome_zarr_path
```

## Load the OME-Zarr file

```python
images = zarr.open(ome_zarr_path, mode="r")
```

# Access images  slice (T, C, Z, Y, X)

```python
channel_0 = images["image_data"][0, 0, 0, :, :].astype(np.float32)
channel_1 = images["image_data"][0, 1, 0, :, :].astype(np.float32)
channel_2 = images["image_data"][0, 2, 0, :, :].astype(np.float32)
channel_3 = images["image_data"][0, 3, 0, :, :].astype(np.float32)

# Display the first image
plt.figure(figsize=(20, 20))
plt.subplot(2, 4, 1)
plt.title("channel_0")
plt.imshow(channel_0, cmap="gray")
plt.axis('off')

plt.subplot(2, 4, 2)
plt.title("channel_1")
plt.imshow(channel_1, cmap="gray")
plt.axis('off')

plt.subplot(1, 4, 1)
plt.title("channel_2")
plt.imshow(channel_2, cmap="gray")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("channel_3")
plt.imshow(channel_3, cmap="gray")
plt.axis('off')
plt.show()
```
