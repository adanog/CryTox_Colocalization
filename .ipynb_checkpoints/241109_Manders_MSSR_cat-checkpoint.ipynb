
# Extended Resolution Analysis with MSSR and Manders Coefficient Calculation

This notebook outlines the process for performing resolution extension using MSSR (Mean Shift Super Resolution microscopy) on fluorescence images, followed by Manders' coefficient analysis to assess colocalization of channels. The workflow is implemented using the **napari-superres** library.

The approach is based on the methodology described in:  
**Torres E. et al., "Extending resolution within a single imaging frame"**  
*Nature Communications, 2022* [doi:10.1038/s41467-022-34693-9](https://www.nature.com/articles/s41467-022-34693-9)

**Script created by Adan Guerrero**  
*Contact: adan.guerrero@ibt.unam.mx*

## Workflow Overview

1. **Resolution Extension (MSSR)**: The raw fluorescence images are processed using the MSSR technique to enhance resolution.
   
2. **Manders' Coefficient Calculation**: Following MSSR, we calculate Manders' coefficients to evaluate the degree of colocalization between channels.

3. **Saving Results**: Results (such as Manders' coefficients) are saved and can be reloaded for further analysis or validation (all_manders_results.pkl)

---

## **Experimental Design**

The experimental design is centered around analyzing the colocalization of toxins with key cellular structures, using various fluorescence channels to track specific cellular components.

Four-channel images were collected using an Olympus FV1000 microscope at a resolution of 60 nm per pixel, with a 60x lens. The first three channels were fluorescence, and the last channel was DIC.

Regions of Interest (5 × 5 µm²) were manually created using the rectangle tool in FIJI ImageJ v1.54f. The ROI collection was saved as a .zip file using the following menu path:
Analyze → Tools → ROI Manager → More → Save.

### **1. Actin Channels**:
   - **C1 (blue)**: DAPI (nucleus)
   - **C2 (cyan)**: Toxin
   - **C3 (red)**: Actin
   - **C4 (gray)**: Additional marker
   
   **Colocalization Comparisons**:
   - **C1 + C2**: Colocalization between the toxin and the nucleus (DAPI).
   - **C2 + C3**: Colocalization between the toxin and actin to observe cytoskeletal interactions.

   The **Lysosome and Mitochondrial channels** are treated similarly to actin for their respective analysis.

### **2. Endoplasmic Reticulum (ER) Channels**:
   - **C1 (blue)**: DAPI (nucleus)
   - **C2 (green)**: ER
   - **C3 (red)**: Toxin
   - **C4 (gray)**: Additional marker

   **Colocalization Comparisons**:
   - **C1 + C3**: Colocalization between the toxin and the nucleus.
   - **C2 + C3**: Colocalization between the ER and the toxin.

### **3. Endosomal Channels**:
   - **C1 (blue)**: DAPI (nucleus)
   - **C2 (cyan)**: Toxin
   - **C3 (gray)**: Additional marker
   - **C4 (green)**: Endosomes

   **Colocalization Comparisons**:
   - **C1 + C2**: Colocalization between the toxin and the nucleus.
   - **C2 + C4**: Colocalization between the toxin and endosomes.

---

### **Additional Experimental Notes**:
- The channels for **Lysosomes**, **Mitochondria**, and **Endosomes** are processed similarly for colocalization analysis but focused on the **cytoplasmic region** only.
- **Nuclear colocalization** is separated into distinct figures for better clarity and interpretation.
- The vesicles corresponding to **ER, Endosomes, Lysosomes**, and **Mitochondria** are also analyzed for colocalization with toxins.

```python
#################### Libraries ################
import os
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tifffile
from skimage import io
from read_roi import read_roi_file
from skimage.filters import gaussian
from napari_superres.core_mssr import mssr_class
mssr_instance = mssr_class()

import seaborn as sns
from scipy.stats import kruskal, ranksums
from itertools import combinations
from statsmodels.stats.multitest import multipletests
```

## Functions for Image Processing
This section includes custom functions for MSSR processing and calculating Manders' coefficients.

```python
#################### Functions ###################
# Define the functions
def list_files(folder_path):
    """List all files in the specified folder."""
    return os.listdir(folder_path)

def extract_base_names_and_zip_files(files):
  """Extract base names from tif files and zip files from the list of files."""
  base_names = {}
  zip_files = {}
  for filename in files:
    if filename.endswith('.tif'):
      initial_number = filename.split('-')[0].split()[-1]
      base_name = os.path.splitext(filename)[0]
      base_names[initial_number] = base_name
    elif filename.endswith('.zip'):
      initial_number = filename.split('-')[1].split()[-1]
      zip_files[initial_number] = filename
  return base_names, zip_files

def create_folder(folder_path, base_name):
    """Create a folder for the given base name."""
    folder_name = os.path.join(folder_path, base_name)
    os.makedirs(folder_name, exist_ok=True)
    return folder_name

def extract_zip_file(folder_path, zip_filename, folder_name):
    """Extract the zip file into the specified folder."""
    with zipfile.ZipFile(os.path.join(folder_path, zip_filename), 'r') as zip_ref:
        zip_ref.extractall(folder_name)
    print(f"Extracted {zip_filename} to {folder_name}")


# Extract the rectangular region from the specified channel
def extract_rect_roi(image, left, top, width, height):
    return image[top:top+height, left:left+width]


# def process_image(image):
#     """Apply MSSR and Gaussian smoothing to the image."""
#     image_smooth = gaussian(image, sigma=blur)
#     image_mssr = sfMSSR(image_smooth, fwhm=8, amp=3, order=1, mesh=True, ftI=False, intNorm=False)
#     return image_mssr

def apply_threshold(image, threshold_value):
    """Apply a simple threshold to the image."""
    return (image > threshold_value).astype(np.uint8)

def manders_colocalization(image1, image2):
    overlap = np.logical_and(image1 > 0, image2 > 0)
    m1 = np.sum(overlap) / np.sum(image1 > 0) if np.sum(image1 > 0) > 0 else 0
    m2 = np.sum(overlap) / np.sum(image2 > 0) if np.sum(image2 > 0) > 0 else 0
    return m1, m2


def normalize_channel(channel, lower_percentile, upper_percentile):
    lower_bound = np.percentile(channel, lower_percentile)
    upper_bound = np.percentile(channel, upper_percentile)
    channel = np.clip(channel, lower_bound, upper_bound)
    # Avoid division by zero
    if upper_bound == lower_bound:
        return np.zeros_like(channel)
    return (channel - lower_bound) / (upper_bound - lower_bound)


def process_roi_files(folder_name, channel_2, channel_3, threshold_value=0.05, visualization_percentiles=(2.5, 97.5), show_plots=True):
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.ndimage import gaussian_filter as gaussian

    # Path to the folder containing .roi files
    roi_folder_path = folder_name

    # List all .roi files in the folder
    roi_files = [f for f in os.listdir(roi_folder_path) if f.endswith('.roi')]

    M = []

    # Process each .roi file
    for roi_file in roi_files:
        roi_path = os.path.join(roi_folder_path, roi_file)
        roi = read_roi_file(roi_path)

        # Extract ROI regions for channels 2 and 3
        roi_info = next(iter(roi.values()))  # Get the first (and only) ROI in the file
        left = roi_info['left']
        top = roi_info['top']
        width = roi_info['width']
        height = roi_info['height']

        roi_channel_2 = extract_rect_roi(channel_2, left, top, width, height)
        roi_channel_3 = extract_rect_roi(channel_3, left, top, width, height)

        # Compute Manders' colocalization coefficients before MSSR
        m1_before, m2_before = manders_colocalization(roi_channel_2, roi_channel_3)

        # Apply Gaussian filter
        roi_channel_2 = gaussian(roi_channel_2, sigma=2)
        roi_channel_3 = gaussian(roi_channel_3, sigma=2)

        # Apply MSSR
        roi_channel_2_mssr = mssr_instance.sfMSSR(img=roi_channel_2, fwhm=fwhm, amp=amp, order=order, mesh=mesh, ftI=ftI, intNorm=intNorm)
        roi_channel_3_mssr = mssr_instance.sfMSSR(img=roi_channel_3, fwhm=fwhm, amp=amp, order=order, mesh=mesh, ftI=ftI, intNorm=False)

        # Apply threshold
        thresh_img1_mssr = apply_threshold(roi_channel_2_mssr, threshold_value)
        thresh_img2_mssr = apply_threshold(roi_channel_3_mssr, threshold_value)

        # Compute Manders' colocalization coefficients after MSSR
        m1_after, m2_after = manders_colocalization(thresh_img1_mssr, thresh_img2_mssr)

        # Append the results as a tuple
        M.append((roi_info['name'], round(m1_before, 3), round(m2_before, 3), round(m1_after, 3), round(m2_after, 3)))
        print(f"Manders' colocalization coefficients before MSSR: M1 = {round(m1_before, 3)}, M2 = {round(m2_before, 3)}")
        print(f"Manders' colocalization coefficients after MSSR: M1 = {round(m1_after, 3)}, M2 = {round(m2_after, 3)}")

        # Normalize channels for visualization
        norm_channel_2 = normalize_channel(roi_channel_2, visualization_percentiles[0], visualization_percentiles[1])
        norm_channel_3 = normalize_channel(roi_channel_3, visualization_percentiles[0], visualization_percentiles[1])
        norm_channel_2_mssr = normalize_channel(roi_channel_2_mssr, visualization_percentiles[0], visualization_percentiles[1])
        norm_channel_3_mssr = normalize_channel(roi_channel_3_mssr, visualization_percentiles[0], visualization_percentiles[1])

        # Create RGB images
        green_channel_2 = np.zeros((roi_channel_2.shape[0], roi_channel_2.shape[1], 3), dtype=float)
        red_channel_3 = np.zeros((roi_channel_3.shape[0], roi_channel_3.shape[1], 3), dtype=float)
        green_channel_2_mssr = np.zeros((roi_channel_2_mssr.shape[0], roi_channel_2_mssr.shape[1], 3), dtype=float)
        red_channel_3_mssr = np.zeros((roi_channel_3_mssr.shape[0], roi_channel_3_mssr.shape[1], 3), dtype=float)

        green_channel_2[..., 1] = norm_channel_2  # Channel 2 in green
        red_channel_3[..., 0] = norm_channel_3    # Channel 3 in red
        green_channel_2_mssr[..., 1] = norm_channel_2_mssr  # Channel 2 in green after MSSR
        red_channel_3_mssr[..., 0] = norm_channel_3_mssr    # Channel 3 in red after MSSR

        if show_plots:
            # Display the results in a single row
            plt.figure(figsize=(20, 5))

            # Display the original images with adjusted brightness and contrast
            plt.subplot(1, 6, 1)
            plt.title(f"Ch 2 before MSSR\n{roi_file}")
            plt.imshow(green_channel_2)
            plt.axis('off')

            plt.subplot(1, 6, 2)
            plt.title(f"Ch 3 before MSSR\n{roi_file}")
            plt.imshow(red_channel_3)
            plt.axis('off')

            # Display the images after MSSR with adjusted brightness and contrast
            plt.subplot(1, 6, 3)
            plt.title(f"Ch 2 after MSSR\n{roi_file}")
            plt.imshow(green_channel_2_mssr)
            plt.axis('off')

            plt.subplot(1, 6, 4)
            plt.title(f"Ch 3 after MSSR\n{roi_file}")
            plt.imshow(red_channel_3_mssr)
            plt.axis('off')

            # Display overlay of both channels before MSSR
            combined_before = np.zeros((roi_channel_2.shape[0], roi_channel_2.shape[1], 3), dtype=float)
            combined_before[..., 1] = norm_channel_2  # Channel 2 in green component
            combined_before[..., 0] = norm_channel_3  # Channel 3 in red component

            plt.subplot(1, 6, 5)
            plt.title(f"Overlay before MSSR\n{roi_file}")
            plt.imshow(combined_before)
            plt.axis('off')

            # Display overlay of both channels after MSSR
            combined_after = np.zeros((roi_channel_2_mssr.shape[0], roi_channel_2_mssr.shape[1], 3), dtype=float)
            combined_after[..., 1] = norm_channel_2_mssr  # Channel 2 in green component
            combined_after[..., 0] = norm_channel_3_mssr  # Channel 3 in red component

            plt.subplot(1, 6, 6)
            plt.title(f"Overlay after MSSR\n{roi_file}")
            plt.imshow(combined_after)
            plt.axis('off')

            plt.show()

    return M
def filter_and_extract_manders_coeff(all_manders_results, order_types, threshold_values):
    filtered_results = {threshold: {order_type: [] for order_type in order_types} for threshold in threshold_values}

    for entry in all_manders_results:
        text, value, data_list = entry
        for order_type in order_types:
            filtered_data = [
                (item[1], item[2], item[3], item[4]) for item in data_list
                if order_type in item[0]
            ]
            if filtered_data:
                for threshold in threshold_values:
                    if value == threshold:
                        filtered_results[threshold][order_type].extend(filtered_data)

    return filtered_results

def plot_boxplots_and_compare_order_types(filtered_results, threshold_values, order_types, output_filename):
    color_blind_palette = ['#0072b2', '#56b4e9', '#009966', '#f0e442', '#e69f03', '#d55e00', '#cc79a7'] # Lighter green and orange color-blind-friendly colors

    n_thresholds = len(threshold_values)
    fig, axes = plt.subplots(1, n_thresholds, figsize=(8, 4), sharey=True)  # Adjusted figsize to 4in x 8in

    if n_thresholds == 1:
        axes = [axes]

    for idx, threshold in enumerate(threshold_values):
        combined_data = []

        for order_type in order_types:
            coeff_data = filtered_results[threshold][order_type]
            if coeff_data:
                for item in coeff_data:
                    combined_data.append((order_type, 'M1_sr', item[2]))
                    combined_data.append((order_type, 'M2_sr', item[3]))

        if combined_data:
            combined_df = pd.DataFrame(combined_data, columns=['OrderType', 'Coefficient', 'Value'])

            sns.boxplot(ax=axes[idx], x='Coefficient', y='Value', hue='OrderType',
                        data=combined_df, palette=color_blind_palette, notch=True)
            axes[idx].set_title(f'Threshold {threshold}')
            axes[idx].set_ylabel('Manders Coefficient' if idx == 0 else '')
            axes[idx].set_ylim(0, 1)  # Set y-axis limits from 0 to 1
            if idx != n_thresholds - 1:
                axes[idx].legend_.remove()

    axes[0].legend(title='Order Type', loc='upper right')
    plt.suptitle('Manders coefficients comparison by order type', y=1.02)
    plt.tight_layout()

    # Save the plot as a PDF with specified size
    fig.savefig(folder_path+"/"+output_filename, format='pdf')

    plt.show()

#################
def significance_asterisks(p_value):
    """
    Returns a string of asterisks corresponding to the significance level.
    """
    if p_value < 0.001:
        return '***'
    elif p_value < 0.01:
        return '**'
    elif p_value < 0.05:
        return '*'
    else:
        return 'ns'

def perform_kruskal_posthoc(threshold_values, order_types, filtered_results, output_filename):
    """
    Performs Kruskal-Wallis test and post hoc pairwise comparisons, and saves the results to a CSV file.

    Args:
        threshold_values (list): List of threshold values.
        order_types (list): List of order types.
        filtered_results (dict): Dictionary containing filtered results.
        output_filename (str): Path to save the CSV file.

    Returns:
        None (Saves adjusted p-values for pairwise comparisons in a CSV file.)
    """
    results_list = []

    for threshold in threshold_values:
        combined_data = []
        for order_type in order_types:
            coeff_data = filtered_results[threshold][order_type]
            if coeff_data:
                for item in coeff_data:
                    combined_data.append((order_type, 'M1_sr', item[2]))
                    combined_data.append((order_type, 'M2_sr', item[3]))

        df = pd.DataFrame(combined_data, columns=['OrderType', 'Metric', 'Value'])

        # Separate data for 'M1_sr' and 'M2_sr'
        data_m1 = df[df['Metric'] == 'M1_sr']
        data_m2 = df[df['Metric'] == 'M2_sr']

        # Mann-Whitney U test for all pairwise comparisons
        p_values_m1 = {}
        p_values_m2 = {}
        for order1, order2 in combinations(df['OrderType'].unique(), 2):
            data1_m1 = data_m1[data_m1['OrderType'] == order1]['Value']
            data2_m1 = data_m1[data_m1['OrderType'] == order2]['Value']
            _, p_value_m1 = ranksums(data1_m1, data2_m1)
            p_values_m1[(order1, order2, 'M1_sr')] = p_value_m1

            data1_m2 = data_m2[data_m2['OrderType'] == order1]['Value']
            data2_m2 = data_m2[data_m2['OrderType'] == order2]['Value']
            _, p_value_m2 = ranksums(data1_m2, data2_m2)
            p_values_m2[(order1, order2, 'M2_sr')] = p_value_m2

        # Adjust p-values (e.g., Bonferroni correction)
        adjusted_p_values_m1 = multipletests(list(p_values_m1.values()), method='bonferroni')[1]
        adjusted_p_values_m2 = multipletests(list(p_values_m2.values()), method='bonferroni')[1]

        # Store results in a list for saving later
        for (order1, order2, metric), adj_p_value in zip(p_values_m1.keys(), adjusted_p_values_m1):
            asterisks = significance_asterisks(adj_p_value)
            results_list.append([threshold, metric, order1, order2, adj_p_value, asterisks])

        for (order1, order2, metric), adj_p_value in zip(p_values_m2.keys(), adjusted_p_values_m2):
            asterisks = significance_asterisks(adj_p_value)
            results_list.append([threshold, metric, order1, order2, adj_p_value, asterisks])

    # Save results to a CSV file
    results_df = pd.DataFrame(results_list, columns=['Threshold', 'Metric', 'Order 1', 'Order 2', 'Adjusted p-value', 'Significance'])
    results_df.to_csv(folder_path+"/"+output_filename, index=False)

    print(results_df)

```

Each experiment is processed using MSSR, and the Manders coefficient is computed afterward

```python

M = []
all_manders_results = []
for initial_number, base_name in base_names.items():
    if initial_number in unzipped_folders:
        folder_name = unzipped_folders[initial_number]
        tif_file = base_names[initial_number] + '.tif'
        tif_path = os.path.join(folder_path, tif_file)
        if os.path.isfile(tif_path):
            print(tif_path)
            # Read the multi-page TIFF file
            images = io.imread(tif_path)

            # Extract channels 2 and 3
            channel_2 = images[:,:,ch_m1].astype(np.float32)
            channel_3 = images[:,:,ch_m2].astype(np.float32)

            # List to hold all results
            #all_manders_results = []
            # Process each threshold value
            for threshold in threshold_values:
                manders_results = process_roi_files(folder_name, channel_2, channel_3, threshold_value=threshold, visualization_percentiles=(2.5, 97.5), show_plots=False)
                all_manders_results.append((base_names[initial_number], threshold, manders_results))

            if len(images) < 2:
                print(f"Not enough channels in {tif_file}")
                continue

```
